```python
from selenium import webdriver
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# 시즌 컬럼 생성
for i in range(0, 56):  
    df_name = f"round_df{i:02d}"  
    
    if i // 10 == 0:
        season = "22-23"
    elif i // 10 == 1:
        season = "21-22"
    elif i // 10 == 2:
        season = "20-21"
    elif i // 10 == 3:
        season = "19-20"
    elif i // 10 == 4:
        season = "18-19"
    elif i // 10 == 5:
        season = "17-18"
    
    # 변수 존재 여부 확인 후 "시즌" 컬럼 추가
    if df_name in globals():
        globals()[df_name]["시즌"] = season
    else:
        pass
```

```python
# 라운드 컬럼 생성
for i in range(0, 56):  
    df_name = f"round_df{i:02d}"  
    
    if i % 10 == 0:
        round = 1
    elif i % 10 == 1:
        round = 2
    elif i % 10 == 2:
        round = 3
    elif i % 10 == 3:
        round = 4
    elif i % 10 == 4:
        round = 5
    elif i % 10 == 5:
        round = 6
    
    # 변수 존재 여부 확인 후 "라운드" 컬럼 추가
    if df_name in globals():
        globals()[df_name]["라운드"] = round
    else:
        pass
```

```python
# 빈 리스트 생성
dfs = []

# 00부터 55까지 반복
for i in range(0, 56):
    # 제외할 번호 리스트
    if i % 10 in [6]:  # 06, 16, 26, 36, 46, 56 제외
        continue
    
    # 데이터프레임 이름 생성
    df_name = f"round_df{i:02d}"  # "round_df00", "round_df01", ..., "round_df55" 형식으로 생성
    
    # 해당 변수 이름이 globals()에 있는지 확인
    if df_name in globals():
        dfs.append(globals()[df_name])  # 해당 데이터프레임을 리스트에 추가

# 데이터프레임들을 세로로 합치기
round_df = pd.concat(dfs, axis=0, ignore_index=True)
```

```python
round_df = pd.concat([round_df, df_loaded], axis=0, ignore_index=True)
```
